```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
print("--- Generamos clientes ---")
# generar manchas o grupos de datos
X,y_secreta = make_blobs(n_samples=300, centers=4, cluster_std=0.6,random_state=0)
# vemos los datos
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1], s=50, color='gray')
plt.title("Clientes (Ingresos vs Edad)")
plt.show()
```

```python
print("\n--- Entrenando a K-Means ---")
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
# modelo.fit(X)
kmeans.fit(X)
etiquetas_predichas = kmeans.predict(X)
centroides = kmeans.cluster_centers_
```

```python
print("\n--- Visualizando los grupos descubiertos ---")
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1], c=etiquetas_predichas, s=50,cmap='viridis')
plt.scatter(centroides[:,0], centroides[:,1], c='red', s=200,alpha=0.7,marker='X', label='Centroides')
plt.title("Clientes agrupados por K-Means")
plt.legend()
plt.show()
```
