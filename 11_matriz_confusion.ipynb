```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print("--- Preparando el modelo ---")
# Cargamos los datos
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2, random_state=42)
# Entrenamos el modelo
modelo = RandomForestClassifier(random_state=42)
modelo.fit(X_train, y_train)
# Hacemos las predicciones
predicciones = modelo.predict(X_test)
print("Predicciones listas.")
```

```python
print("\n--- Matriz de Confusion ---")
# Generamos la matriz
matriz = confusion_matrix(y_test, predicciones)
# Le damos formato
plt.figure(figsize=(6,4))
sns.heatmap(matriz, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Maligno (0)', 'Benigno (1)'],
            yticklabels=['Maligno (0)', 'Benigno (1)'])
plt.ylabel('Realidad (lo que tiene el paciente)')
plt.xlabel('Prediccion (lo que dice el modelo)')
plt.title('Matriz de confusion')
plt.show()
```

```python
print("\n--- El reporte de clasificacion ---")
reporte = classification_report(y_test, predicciones, target_names=['Maligno (0)','Benigno (1)'])
print(reporte)
```
