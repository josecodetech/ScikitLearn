```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
print("--- Preparando los datos ---")
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
print("--- Configurando el modelo ---")
bosque = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
print("--- Validacion cruzada ---")
notas_cv = cross_val_score(bosque, X, y, cv=5, n_jobs=-1)
print("\nExamenes terminados!")
print(f"Las 5 notas obtenidas son: {notas_cv}")
```

```python
# Calculamos la media y desviacion estandar (el margen de error)
nota_media = np.mean(notas_cv)
desviacion = np.std(notas_cv)
print("\n--- Resultado final oficial ---")
print(f"Precision Media: {nota_media*100:.2f}%")
print(f"Margen de error (Varianza): +/- {desviacion*100:.2f}%")
```
