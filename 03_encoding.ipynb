```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

# 1. Creamos nuestros datos de juguete (con texto)

```python
datos_coches = pd.DataFrame({
    'Precio':[20000,25000,22000,30000],
    'Marca':['Toyota','Ford','Seat','BMW']
})
print("Datos originales:\n",datos_coches)
```

## Seleccionamos SOLO la columna problematica

```python
columna_marca = datos_coches[['Marca']]
```

## Instanciamos la herramienta

```python
codificador = OneHotEncoder(sparse_output=False)
```

## Paso magico: fit y transform a la vez

```python
marcas_codificadas = codificador.fit_transform(columna_marca)
print("\nMatriz matematica resultante:\n", marcas_codificadas)
```

## Pedimos nombres de nuevas columnas

```python
nombres_nuevas = codificador.get_feature_names_out(['Marca'])
```

## Convertimos la matriz en un DataFrame

```python
df_marcas = pd.DataFrame(marcas_codificadas, columns=nombres_nuevas)
```

## Lo pegamos al dataframe original

```python
dataset_final = pd.concat([datos_coches[['Precio']], df_marcas],axis=1)
print("\n--- DATOS FINALES ---")
print(dataset_final)
```
