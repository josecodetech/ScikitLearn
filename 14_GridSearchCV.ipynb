```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
print("--- Preparando el terreno ---")
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
print("--- Definiendo la cuadricula (diccionario de opciones) ---")
parametros_a_probar = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 5, 10],
}
```

```python
print("--- Configurando la busqueda magica (GridSearchCV) ---")
bosque_base = RandomForestClassifier(random_state=42)
```

```python
# Instanciamos el buscador
# cv = 5
buscador = GridSearchCV(estimator=bosque_base,
                        param_grid=parametros_a_probar,
                        cv = 5,
                        n_jobs=-1)
print("\nBuscando el mejor modelo...(Esto puede tardar...)")
buscador.fit(X,y)
print("Busqueda terminada")
```

```python
print("\n --- Resultados del campeonato ---")
print(f"La mejor combinacion de hiperparametros es: {buscador.best_params_}")
print(f"La nota media perfecta (Accuracy) conseguida fue: {buscador.best_score_*100:.2f}%")
mejor_modelo_del_mundo = buscador.best_estimator_
```
