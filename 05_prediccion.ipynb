# Paso 1: Importar y cargar datos

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
X = pd.DataFrame({'Horas_Estudio':[1,2,3,4,5]})
y = pd.Series([2,4,6,8,10])
```

# Paso 2: Instanciar 

```python
modelo = LinearRegression()
```

# Paso 3: Entrenar

```python
modelo.fit(X,y)
print("El modelo ha aprendido la formula secreta.")
```

# Paso 4: Predecir

```python
horas_nuevas = pd.DataFrame({'Horas_Estudio':[7,8.5]})
predicciones = modelo.predict(horas_nuevas)
print("\nPredicciones:")
for horas, nota in zip(horas_nuevas['Horas_Estudio'],predicciones):
    print(f"Si estudias {horas} horas, sacaras un: {nota}")
```

# Variables internas del modelo


```python
modelo.coef_
```

```python
print(modelo.score(X,y))
```
