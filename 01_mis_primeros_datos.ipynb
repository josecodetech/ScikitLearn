```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=100, n_features = 1, noise = 15, random_state = 10)
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X, y, color='blue', alpha=0.6)
plt.title("Nuestros primeros datos sinteticos")
plt.xlabel("Tamaño de la casa")
plt.ylabel("Precio")
plt.grid(True)
plt.show()
```
