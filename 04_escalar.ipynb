```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

# Datos originales

```python
df = pd.DataFrame({
    'Edad':[20,30,40,50,60],
    'Salario':[20000,40000,60000,80000,100000]
})
print("Datos originales:\n", df)
```

# Aplicando StandardScaler

```python
escalador = StandardScaler()
```

```python
datos_escalados = escalador.fit_transform(df)
```

```python
df_escalado = pd.DataFrame(datos_escalados, columns=['Edad','Salario'])
print("\nDatos escalados:\n", df_escalado)
```
